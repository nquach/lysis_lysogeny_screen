```python
%matplotlib inline
import matplotlib
import numpy as np 
import os
import matplotlib.pyplot as plt
import cPickle as pickle
from SLIP_functions import classify_well_gmm2
from keio_names import get_keio_names, pos_to_strain
import seaborn as sns
import json
from scipy.stats import beta
import utils
```

```python
BLUE = (0, 114.0/255.0, 178.0/255.0)
VERM = (213.0/255.0, 94.0/255.0, 0)
GREEN = (0, 158.0/255.0, 115.0/255.0)
PURPLE = (204.0/255.0, 121.0/255.0, 167.0/255.0)
BLACK = (0,0,0)

colors = [BLUE, BLACK, VERM, GREEN, PURPLE]
```

```python
def calculate_MOI(frac_infected):
	Nc = 100
	N_infected = 100*frac_infected
	x = np.linspace(0,5,200)
	gamma = np.float(Nc)*np.log(1-1/np.float(Nc))
	posterior = np.abs(gamma*np.exp(gamma*x))*beta.pdf(np.exp(gamma*x), 1+Nc-N_infected, 1+N_infected)
	idx = np.argmax(posterior)
	return x[idx]
```

```python
root_direc = utils.ROOT_DIREC
direc = os.path.join(root_direc, 'pkl_files')
classification_wells = ['H12']
plate_numbers = ['CS_duplicate1', 'CS_duplicate2']
row_data = ['A', 'B', 'C', 'D', 'E']
col_data = range(1,13)
pooled_FITC = {}
pooled_cherry = {}
```

```python
wells = []
for row in row_data:
	for col in col_data:
		well = row + str(col)
		wells += [well]

for plate_number in plate_numbers:
	root_direc = os.path.join(direc, plate_number)
	mean_FITC_name = os.path.join(root_direc, 'mean_FITC.pkl')
	mean_cherry_name = os.path.join(root_direc, 'mean_cherry.pkl')
	mean_FITC = pickle.load(open(mean_FITC_name, 'rb'))
	mean_cherry = pickle.load(open(mean_cherry_name, 'rb'))
	for well in wells:
		if well in pooled_FITC.keys():
			pooled_FITC[well] = pooled_FITC[well] + mean_FITC[well]
			pooled_cherry[well] = pooled_cherry[well] + mean_cherry[well]
		else:
			pooled_FITC[well] = mean_FITC[well]
			pooled_cherry[well] = mean_cherry[well]
```

```python
pooled_FITC2 = {}
pooled_cherry2 = {}
carbon_source = ['Fructose','Glycerol','Glucose','Maltose','Pyruvate']
index = 0
fru_mean = []
fru_var = []
gly_mean = []
gly_var = []
glu_mean = []
glu_var = []
mal_mean = []
mal_var = []
pyr_mean = []
pyr_var = []
for carbon in carbon_source:
	pooled_FITC2[carbon] = []
	pooled_cherry2[carbon] = []
	for i in range(0,12,2):
		#print index
		pos1 = row_data[index] + str(col_data[i])
		pos2 = row_data[index] + str(col_data[i+1])

		pooled_FITC2[carbon] = pooled_FITC2[carbon] + [pooled_FITC[pos1] + pooled_FITC[pos2]]
		pooled_cherry2[carbon] = pooled_cherry2[carbon] + [pooled_cherry[pos1] + pooled_cherry[pos2]]
	index = index + 1
```

```python
mean_dict = {}
var_dict = {}
frac_inf_dict = {}
for carbon in carbon_source:
	print carbon
	mean_list = []
	var_list = []
	frac_inf_list = []
	for i in range(6):
		lytic_list, lysogenic_list, uninfected_list = classify_well_gmm2(pooled_FITC2[carbon][i], pooled_cherry2[carbon][i], classification_wells = classification_wells)
		n_lytic = len(lytic_list)
		n_lyso = len(lysogenic_list)
		n_uninfected = len(uninfected_list)
		infected = n_lytic + n_lyso
		total = infected + n_uninfected
		frac_inf = 0
		if total > 0:
			frac_inf = float(infected)/float(total)
		lysis_ratio = 0
		if infected > 0:
			lysis_ratio = float(n_lytic)/float(infected)
		else:
			lysis_ratio = 0.5
		a = float(1+n_lytic)
		b = float(1+n_lyso)
		var = a*b/(np.square(a+b)*(a+b+1))
		mean_list.append(lysis_ratio)
		var_list.append(var)
		frac_inf_list.append(frac_inf)
	mean_dict[carbon] = mean_list
	var_dict[carbon] = var_list
	frac_inf_dict[carbon] = frac_inf_list
```

```python
#Calculate the MOI for each condition
MOI_dict = {}
for source in frac_inf_dict.keys():
	FI_list = frac_inf_dict[source]
	MOI_list = []
	for FI in FI_list:
		MOI = calculate_MOI(FI)
		MOI_list.append(MOI)
	MOI_dict[source] = MOI_list
```

```python
#Select mode for x axis. 'MOI' or 'vol' or 'FI'
mode = 'vol'
#Select if you want to print cell fate stats
verbose = False
```
